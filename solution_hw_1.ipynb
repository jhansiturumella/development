```python
# Importing the dataset
column_names= ['ID','title','year','rating','votes','duration','genres']
df=pd.read_csv('movie_data.txt',header= None, names=column_names, sep='\t')
```

# Cleaning

```python
# checking the missing value
df.isnull().sum()
```

```python
# droping the missing value
df_1=df.dropna()
```

```python
df_1.isnull().sum()
```

```python
# removing the min str from the duration column
df_1['duration']=df_1['duration'].str.slice(0,-6)  #it takes(start,step,stop)
```

```python
df_1.head()
```

```python
# Converting duration column minute to second
df_1['duration']=df_1['duration'].apply(lambda x: int(x)*60)
```

```python
df_1.head()
```

```python
# removing the year str from the title column
df_1['title']=df_1['title'].str.slice(0,-7)
```

```python
df_1.head()
```

# Concatenating

```python
# Importing the movie_crew dataset
movie_crew_df= pd.read_csv('crew_data.tsv.gz', sep='\t')
```

```python
movie_crew_df.head()
```

```python
# rename the column name
movie_crew_df=movie_crew_df.rename(columns = {'tconst':'ID', 'directors':'director_ids'})
```

```python
# setting the index
movie_crew_df.set_index('ID',inplace=True)
```

```python
  movie_crew_df.head()
```

```python
# removing the writers column
del movie_crew_df['writers']
```

```python
movie_crew_df.head()
```

```python
#  Importing person data
```

```python
person_data=pd.read_csv('name_data.tsv.gz', sep='\t')
```

```python
person_data.head()
```

```python
person_data=person_data.rename(columns = {'nconst':'Person_ID', 'primaryName':'name'})
```

```python
#person_data.head()
```

```python
person_data.set_index('Person_ID',inplace=True)
```

```python
person_data.head()
```

```python
person_data=person_data.drop(person_data.columns[1:],axis=1)
```

```python
person_data.head()
```

```python
# reset the index
person_data=person_data.reset_index()
```

```python
person_data.head()
```

```python
# merging movie_crew_df
df_new= df_1.copy()
df_1=df_1.drop(df_1.columns[1:], axis=1)
#merge_movie_crew= pd.concat([df_1,movie_crew_df])
merge_movie_crew=pd.merge(df_1, movie_crew_df, on='ID', how='outer')
```

```python
merge_movie_crew.isnull().sum()
```

```python
# droping the null value
merge_movie_df=merge_movie_crew.dropna()
```

```python
merge_movie_df.isnull().sum()
```

```python
#Remove movies with more than one director
df_movie= merge_movie_df[merge_movie_df["director_ids"].str.len()==9]
```

```python
df_movie=df_movie.rename(columns={'director_ids':'director_id'})
```

```python
df_movie.head()
```

```python
df_movie=df_movie.reset_index()
```

```python
df_movie.head()
```

```python
merge_person=pd.merge(df_movie, person_data, left_on='director_id',right_on="Person_ID", how='inner')
```

```python
merge_person.head()
```

```python
merge_person=merge_person.drop(merge_person.columns[1:4],axis=1)
```

```python
merge_person.head()
```

```python
# rename the column name
merge_person=merge_person.rename(columns={'name':'director'})
```
