### Step 1. Import the necessary libraries

```python
## Importng necessary libraries
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep = '\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
## Saving for temp variable cnt
cnt = chipo[chipo.item_price > '$10.00']
print(cnt.count())
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
## Alternatively we can use df[[item_name,item_price]]
print(chipo.loc[:,['item_name', 'item_price']])
```

### Step 6. Sort by the name of the item

```python
print(chipo.sort_values('item_name').head(20))
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
df = chipo[chipo.item_price == chipo.item_price.max()]
maximum = df.groupby(['item_price', 'quantity'])
maximum.first()  ## Used only for groupby
## ALtenatively df.iloc[0],df.head(1)
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
veggie = chipo[chipo.item_name == 'Veggie Salad Bowl']
print(veggie.shape[0])
    
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
canned = chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)]
print('No. of times Canned Soda has been ordered is {}'.format(canned.shape[0]))
```
